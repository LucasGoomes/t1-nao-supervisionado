```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import cluster
from collections import Counter
import numpy as np
```

### Original data

```python
df_cb_original = pd.read_csv('data/cleaned_data/CB_original.csv')

df_cm_original = pd.read_csv('data/cleaned_data/CM_original.csv')

df_lb_original = pd.read_csv('data/cleaned_data/LB_original.csv')
```

### Prepared training data

```python
df_cb = pd.read_csv('data/prepared_data_2/by_position/kmeans/CB.csv')
df_cb_players = pd.read_csv('data/prepared_data_2/by_position/kmeans/CB_names.csv')

df_cm = pd.read_csv('data/prepared_data_2/by_position/kmeans/CM.csv')
df_cm_players = pd.read_csv('data/prepared_data_2/by_position/kmeans/CM_names.csv')

df_lb = pd.read_csv('data/prepared_data_2/by_position/kmeans/LB.csv')
df_lb_players = pd.read_csv('data/prepared_data/by_position/kmeans/LB_names.csv')
```

## CB Position

```python
k_means_cb = cluster.KMeans(n_clusters=2, max_iter=50, init='random',n_init=25, random_state=123)
k_means_cb.fit(df_cb)
```

```python
labels = k_means_cb.labels_
df_cb_original['cluster'] = labels
df_cb_original['Player Name'] = df_cb_players
```

```python
df_cb_original.head(2)
df_cb_original.to_csv(f'data/final_data/DB_final.csv', index=False)  
```

```python
df_cb_original_numeric = df_cb_original.drop('Player Name', axis=1) 

cluster_means = df_cb_original_numeric.groupby('cluster').mean()  
cluster_means.head()
```

```python
cluster_medians = df_cb_original_numeric.groupby('cluster').median()  
cluster_medians.head()
```

## CM Position

```python
k_means_cm = cluster.KMeans(n_clusters=2, max_iter=50, init='random',n_init=25, random_state=123)
k_means_cm.fit(df_cm)
```

```python
labels = k_means_cm.labels_
df_cm_original['cluster'] = labels
df_cm_original['Player Name'] = df_cm_players
```

```python
df_cm_original.head(2)
df_cm_original.to_csv(f'data/final_data/CM_final.csv', index=False)  
```

```python
df_cm_original_numeric = df_cm_original.drop('Player Name', axis=1) 

cluster_means = df_cm_original_numeric.groupby('cluster').mean()  
cluster_means.head()
```

```python
cluster_medians = df_cm_original_numeric.groupby('cluster').median()  
cluster_medians.head()
```
