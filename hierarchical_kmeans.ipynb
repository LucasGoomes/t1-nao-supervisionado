```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster
import os
from sklearn.cluster import AgglomerativeClustering  
from scipy.cluster.hierarchy import dendrogram
```

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
folder_path = 'data/prepared_data/by_position/kmeans/'  
files = os.listdir(folder_path)  

file_pairs = [(files[i], files[i + 1]) for i in range(0, len(files), 2)]

dfs = {} 
models = {} 
```

```python
file_pairs
```

```python
for file, player_names_file in file_pairs:  
    df = pd.read_csv(folder_path + file)  
    player_names = pd.read_csv(folder_path + player_names_file)  

    cluster = AgglomerativeClustering(n_clusters=None, metric='euclidean', linkage='single', distance_threshold=0)  
    df['cluster'] = cluster.fit_predict(df)  

    df['player_name'] = player_names  
    
    dfs[file] = df  
  
    models[file] = cluster  
```

```python
plt.title('Dendrogram for CB.csv')  
plot_dendrogram(models['CB.csv'], truncate_mode='level', p=3)  
plt.show()  
```
