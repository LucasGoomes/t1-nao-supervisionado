```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import cluster
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples, silhouette_score
import os
```

```python
folder_path = 'data/prepared_data/by_position/kmeans/'  
files = os.listdir(folder_path)  

file_pairs = [(files[i], files[i + 1]) for i in range(0, len(files), 2)]

dfs = {} 
models = {} 

numClusters = np.arange(1, 7, 1) 
numClustersSil = np.arange(2, 7, 1)  

sse = {}
sc = {}
```

```python
file_pairs
```

```python
for file, player_names_file in file_pairs:  
    df = pd.read_csv(folder_path + file)  
    player_names = pd.read_csv(folder_path + player_names_file)  
  
    sse[file] = []
  
    for k in numClusters:  
        k_means = KMeans(n_clusters=k, max_iter=50,init='random', n_init=25, random_state=123)  
        k_means.fit(df)  
        sse[file].append(k_means.inertia_)  
  
    df['player_name'] = player_names  
  
    dfs[file] = df  
```

```python
file = 'CB.csv'
plt.plot(numClusters, sse[file])  
plt.xlabel('Number of Clusters')  
plt.ylabel('SSE')  
plt.title(f'Elbow plot for {file}')  
plt.show()  
```
